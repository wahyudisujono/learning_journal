```python
nba=df.copy()
nba.rename(columns={'game_location':'loc','game_result':'result'},inplace=True)
```

## **Drop Colomn**

```python
nba.drop(axis=1, labels=['elo_i','elo_n'], inplace=True)
```

```python
nba
```

## **New Colomn**

```python
nba['selisih']=nba['pts']-nba['opp_pts']
```

## **Pindah Colomn**

```python
nba.columns
```

```python
nba.columns=['gameorder', 'game_id', 'lg_id', '_iscopy', 'year_id', 'date_game',
       'seasongame', 'is_playoffs', 'team_id', 'fran_id', 'pts', 'win_equiv',
       'opp_id', 'opp_fran', 'opp_pts', 'opp_elo_i', 'opp_elo_n', 'loc',
       'result', 'forecast', 'notes','selisih']
```
